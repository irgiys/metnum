## 1. Import library pandas

```python
import pandas as pd
```

## 2. Simpan data dalam sebuah list

```python
fruits = ["apel", "mangga", "durian", "anggur"]
quantities = [20,30,12,40]
```

## 3. Inialisasi data 

```python
S = pd.Series(quantities, index=fruits, name="Buah")
S.plot()
```

```python
S.plot(kind="bar")
```

```python
S.plot.pie(figsize=(6,6))
```
