### 1. Import Library yang diperlukan

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 2. Inisialisasi data dan rumus 

```python
x = np.arange(-3,4)
y = 2 * x + 4
```

### 3. Berikan aturan pada graf dan tampilkan graf

```python
plt.figure(figsize=(6,6))
plt.plot(x,y)
plt.title('Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
```
