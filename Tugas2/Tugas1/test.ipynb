```python
import matplotlib.pyplot as plt

def f(x):
    return 4.15*x**5-2.23*x**3-6.35

def bisection(a, b):
    e = 10**-4
    n = 100
    counter = 1
    condition = True
    while condition:
        c = (a + b) / 2
        print('Iterasi ke-%d,\tC = %0.4f dan f(C) = %0.4f' % (counter, c, f(c)))
        if f(a) * f(c) < 0:
            b = c
        else:
            a = c
        counter += 1
        condition = abs(f(c)) > e and counter <= n
    print('\nDitemukan, akarnya adalah: %0.4f' % c)
    plt.figure(figsize=(6,6))
    plt.plot(f(c),c)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid()

a = float(1)
b = float(2)

if f(a) * f(b) > 0.0:
    print('Nilai a dan b bukan diantara akar.')
else:
   bisection(a, b)
```
